## Cropping

```python
flowers = cv2.imread("../images/flower.png")
flowers = flowers[:,:,::-1]
croppedimg = flowers[400:800, 400:1500]

plt.subplot(1, 2, 1)
plt.imshow(flowers)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(croppedimg)
plt.title('Cropped Image')
```

Cropping is really strightfprward. It involves indexing into existing image and extracting the desired region

## Resizing images

### Method 1: specifying scaling factor fx and fy

```python
resized_cropped_img1 = cv2.resize(croppedimg, None, fx=2, fy=2)
resized_cropped_img2 = cv2.resize(croppedimg, None, fx=4, fy=4)

plt.subplot(1,2,1)
plt.imshow(resized_cropped_img1)

plt.subplot(1,2,2)
plt.imshow(resized_cropped_img2)
```

### Method 2: Specifying the exact size of the output image

```python
width = 2200
height = 800
dim = (width, height)

resized_cropped_img_with_dimentions = cv2.resize(croppedimg, dsize=dim)
plt.imshow(resized_cropped_img_with_dimentions)
```

```python
plt.imshow(cv2.resize(croppedimg, (4400, 1600)))
```

## Resizing while maintaining aspect ration

```python
desiredWidth = 1000

aspectRatio = desiredWidth/croppedimg.shape[1]

desiredHight = int(croppedimg.shape[0] * aspectRatio)
dim = (desiredWidth, desiredHight)

resizedCroppedRegion = cv2.resize(croppedimg, dsize=dim)
plt.imshow(resizedCroppedRegion)
```

```python
def resize_and_show_image(width):
    desiredWidth = width

    aspectRatio = desiredWidth/croppedimg.shape[1]
    
    desiredHight = int(croppedimg.shape[0] * aspectRatio)
    dim = (desiredWidth, desiredHight)
    
    resizedCroppedRegion = cv2.resize(croppedimg, dsize=dim)
    plt.imshow(resizedCroppedRegion)
```

```python
resize_and_show_image(500)
```

## Save and show the images to and from the disk

```python
cv2.imwrite("ResizedCroppedImg.png", resizedCroppedRegion[:,:,::-1])

Image(filename="ResizedCroppedImg.png")
```
