# Annoting images
- Draw lines
- Draw Circles
- Draw rectangles
- Add Text

### Add lines

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import Image
```

```python
image = cv2.imread("falcon9.jpg")
image = image[:,:,::-1]
plt.imshow(image)
```

## Adding Lines

```python
imgline = image.copy()

cv2.line(imgline, (100, 400), (400, 400), (0,2225,0), thickness=5, lineType=cv2.LINE_AA)

plt.imshow(imgline)
```

## Adding Circles

```python
imgcircle = image.copy()

cv2.circle(imgcircle, (610,230), 100, (0,225,0), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imgcircle)
```

## Adding Rectangles

```python
imgrectangle = image.copy()

cv2.rectangle(imgrectangle, (210,10), (350,300), (0,225,0), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imgrectangle)
```

## Adding text to images

```python
imgtext = image.copy()

text = "Falcon 9"
fontcolor = (0,225,0)
fontthickness = 2
fontscale = 1.3
fontface = cv2.FONT_ITALIC

cv2.putText(imgtext, text, (500,100), fontface, fontscale, fontcolor, fontthickness)
plt.imshow(imgtext)
```
